```python
df=x.drop('Dates',axis=1)
df["column_id"] = 0
span = 10

features_rec=pd.DataFrame()

res_df_max=pd.DataFrame()
res_df_min=pd.DataFrame()
res_df_mean=pd.DataFrame()
res_df_median=pd.DataFrame()
res_df_std=pd.DataFrame()
res_df_var=pd.DataFrame()
res_df_lcd005=pd.DataFrame()
res_df_lcd01=pd.DataFrame()
res_df_lcd02=pd.DataFrame()

for i in range(span-1,len(df)):
    df_cut=df[i-span+1:i+1]
    #calculate the 6 features firstly.
    temp_df_max=pd.DataFrame(df_cut.mean().astype(float)).T
    temp_df_min=pd.DataFrame(df_cut.min().astype(float)).T
    temp_df_mean=pd.DataFrame(df_cut.mean().astype(float)).T
    temp_df_median=pd.DataFrame(df_cut.median().astype(float)).T
    temp_df_std=pd.DataFrame(df_cut.std().astype(float)).T
    temp_df_var=pd.DataFrame(df_cut.var().astype(float)).T
    # record the features
    res_df_max=pd.concat([res_df_max,temp_df_max],axis=0)
    res_df_min=pd.concat([res_df_min,temp_df_min],axis=0)
    res_df_mean=pd.concat([res_df_mean,temp_df_mean],axis=0)
    res_df_median=pd.concat([res_df_median,temp_df_median],axis=0)
    res_df_std=pd.concat([res_df_std,temp_df_std],axis=0)
    res_df_var=pd.concat([res_df_var,temp_df_var],axis=0)
    #calculate the large standard deviation where r=0.05,0.1 and 0.2, record the results
    temp_df_lcd005=pd.DataFrame(np.where(temp_df_std>0.05*(temp_df_max-temp_df_min),1,0))
    res_df_lcd005=pd.concat([res_df_lcd005,temp_df_lcd005],axis=0)
    temp_df_lcd01=pd.DataFrame(np.where(temp_df_std>0.1*(temp_df_max-temp_df_min),1,0))
    res_df_lcd01=pd.concat([res_df_lcd01,temp_df_lcd01],axis=0)
    temp_df_lcd02=pd.DataFrame(np.where(temp_df_std>0.2*(temp_df_max-temp_df_min),1,0))
    res_df_lcd02=pd.concat([res_df_lcd02,temp_df_lcd02],axis=0)

max_col_index = [col + f'_max_{span}' for col in df.columns]    
min_col_index = [col + f'_min_{span}' for col in df.columns]
mean_col_index = [col + f'_mean_{span}' for col in df.columns]
median_col_index = [col +f'_median_{span}' for col in df.columns]
std_col_index = [col + f'_std_{span}' for col in df.columns]
var_col_index = [col + f'_var_{span}' for col in df.columns]
lsd005_col_index = [col + f'_large_standard_deviation_'+str(0.05)+'_{span}' for col in df.columns]
lsd01_col_index = [col + f'_large_standard_deviation_'+str(0.1)+'_{span}' for col in df.columns]
lsd02_col_index = [col +f'_large_standard_deviation_'+str(0.2)+'_{span}' for col in df.columns]



res_df_max.columns=max_col_index 
res_df_min.columns=min_col_index
res_df_mean.columns=mean_col_index
res_df_median.columns=median_col_index
res_df_std.columns=std_col_index
res_df_var.columns=var_col_index
res_df_lcd005.columns=lsd005_col_index
res_df_lcd01.columns=lsd01_col_index
res_df_lcd02.columns=lsd02_col_index


for i in range(res_df_max.shape[1]):
    temp_df = pd.concat([res_df_max.iloc[:,i], res_df_min.iloc[:,i], res_df_mean.iloc[:,i],res_df_median.iloc[:,i], res_df_std.iloc[:,i], res_df_var.iloc[:,i],res_df_lcd005.iloc[:,i], res_df_lcd01.iloc[:,i], res_df_lcd02.iloc[:,i]], axis=1)
    features_rec=pd.concat([features_rec, temp_df], axis=1)



features_rec.index=range(span-1,span-1+len(features_rec))
features_rec.to_csv(f'features_{span}.csv',index=True)
```
