train_Xの市場データと経済指標を分離する

```python
train_mkt_data=train_X['市場データ_final']
train_ind_data=train_X['経済指標データ_final']
```

市場データにおいて、数値とインデックスを分ける

```python
temp=train_mkt_data.iloc[6:]
temp.columns=list(train_mkt_data.iloc[5])
temp
```

要らないデータの削除

```python
#金利3年を削除リストに
eliminate_list = ("USGG3YRIndex","USYC2Y3YIndex","USYC3Y5YIndex","USYC3Y7YIndex","USYC3Y10Index","USYC3Y20Index","USYC3Y30Index","BF020305Index","BF020307Index","BF020310Index","BF020320Index","BF020330Index","BF030507Index","BF030510Index","BF030520Index","BF030530Index","BF030710Index","BF030720Index","BF030730Index","BF031020Index","BF031030Index")
eliminate_num_list = []
for j in range(len(train_mkt_data.columns)):
    if train_mkt_data.iloc[5,j] in eliminate_list:
        eliminate_num_list.append('Unnamed: '+str(j))
```

```python
#全部Nanの列を削除リストに：もしこの列のすべての値はNanのならこの列を削除する
for i in range(len(temp.columns)):
    if temp.isnull().all()[i] == True:
        eliminate_num_list.append('Unnamed: '+str(i))
eliminate_num_list
```

```python
train_mkt_data=train_mkt_data.drop(eliminate_num_list,axis=1)
train_mkt_data
```

経済指標データに対する点検を行う

```python
temp2 = train_ind_data.loc[6:]
temp2.columns = list(train_ind_data.iloc[5])
temp2
```

市場データのインデックスを処理する
処理後：各指標名は縦に並んで、指標名は各行のインデックスになり、指標名の属性は各列のインデックスになる。このdfは各指標名の具体的な内容を説明する

```python
train_mkt_info=train_mkt_data.loc[:5].T.copy() #この時、各指標の名前は縦方向に並んで、縦軸のインデックスはunnamed0～107
train_mkt_info.columns=list(train_mkt_info.iloc[0])
train_mkt_info.drop(train_mkt_info.index[0],inplace=True)
train_mkt_info.index=list(train_mkt_info['Dates'])#字母名前で表すの指標名を今行のインデックスにする
train_mkt_info.drop(columns="Dates", inplace=True)#dfから字母指標名を削除(もうインデックスになったため）
train_mkt_info.rename(columns = {np.nan:"補足"}, inplace=True)
train_mkt_info
```

経済指標データのインデックスも同様に処理する

```python
train_ind_info = train_ind_data.loc[:5].T.copy()
train_ind_info.columns = list(train_ind_info.iloc[0])
train_ind_info.drop(train_ind_info.index[0], inplace=True)
train_ind_info.index = list(train_ind_info["Dates"])
train_ind_info.drop(columns="Dates", inplace=True)
train_ind_info.rename(columns = {np.nan:"補足"}, inplace=True)
train_ind_info
```

指標データの値をdfとして抽出し、字母の指標名を列のインデックスにして、行のインデックスを０からと再設定する

```python
train_mkt_data = train_mkt_data.loc[5:].copy()
train_mkt_data.columns = list(train_mkt_data.iloc[0])
train_mkt_data.drop(index=5, inplace=True)
train_mkt_data.reset_index(inplace=True, drop=True)
train_mkt_data.head(5)
```

```python
train_mkt_data.columns
```

経済指標データに対しても同様に処理する

```python
train_ind_data =train_ind_data.loc[5:].copy()
train_ind_data.columns = list(train_ind_data.iloc[0])
train_ind_data.drop(index=5, inplace=True)
train_ind_data.reset_index(inplace=True, drop=True)
train_ind_data.head(5)
```

三つdfの結合：まずは市場データと経済指標データのDates列の様式をdatetime64に変更し、これで三つdfのこのDates列が一致になり、そしてこの列を基準に三つdfを一つに結合する。新dfの各列は：ｙの値、市場データ、経済指標データの順番 

```python
train_df=train_y['Sheet1']
train_mkt_data['Dates']=train_mkt_data['Dates'].astype('datetime64[ns]')
train_ind_data['Dates'] = train_ind_data['Dates'].astype('datetime64[ns]')
train_df = pd.merge(train_df, train_mkt_data, on='Dates')
train_df = pd.merge(train_df, train_ind_data, on='Dates')
train_df
```

変更したものを保存する

```python
train_mkt_data.to_csv('train_mkt_data.csv')
train_ind_data.to_csv('train_ind_data.csv')
train_mkt_info.to_csv('train_mkt_info.csv')
train_ind_info.to_csv('train_ind_info.csv')
train_df.to_csv('train_yX.csv',index=False)
'''
説明：
train_mkt_infoは全ての市場インデックスをふくむ
train_ind_infoは全ての経済インデックスを含む
train＿ｙXはtrainデータで、中には日付、dec_rate、各市場データと経済指標データを含む。
'''
```

validデータの処理

validも指標データの値をdfとして抽出し、字母の指標名を列のインデックスにして、行のインデックスを０からと再設定する

```python
valid_mkt_data=valid_X['市場データ_final'].drop(eliminate_num_list,axis=1)
valid_mkt_data = valid_mkt_data.loc[5:].copy()
valid_mkt_data.columns = list(valid_mkt_data.iloc[0])
valid_mkt_data.drop(index=5, inplace=True)
valid_mkt_data.reset_index(inplace=True, drop=True)
valid_mkt_data.head(5)
```

```python
valid_ind_data = valid_X['経済指標データ_final'].loc[5:].copy()
valid_ind_data.columns = list(valid_ind_data.iloc[0])
valid_ind_data.drop(index=5, inplace=True)
valid_ind_data.reset_index(inplace=True, drop=True)
valid_ind_data.head(5)
```

validデータも三つdfを合併する

```python
valid_df = valid_y['Sheet1']
valid_mkt_data['Dates'] = valid_mkt_data['Dates'].astype('datetime64[ns]')
valid_ind_data['Dates'] = valid_ind_data['Dates'].astype('datetime64[ns]')
valid_df = pd.merge(valid_df, valid_mkt_data, on='Dates')
valid_df = pd.merge(valid_df, valid_ind_data, on='Dates')
valid_df
```

validデータの保存

```python
valid_df.to_csv('valid_yX.csv',index=False)
'''
説明：valid＿ｙXはvalidデータで、中には日付、dec_rate、各市場データと経済指標データを含む。
'''
```
